```python
import pandas as pd

import requests
from bs4 import BeautifulSoup

import unidecode
import re

import time
```

```python
nomes = pd.read_excel("../data_raw/DTB_BRASIL_MUNICIPIO.xls")
nomes = nomes[["cod_ibge", "uf_sigla", "nome_ibge"]]
nomes.head()
```

```python
nomes['mun_path'] = nomes.nome_ibge.apply(unidecode.unidecode).str.replace('\'', '').str.lower().str.replace(' ', '-')
```

```python
web_paths = "https://cidades.ibge.gov.br/brasil/" + nomes['uf_sigla'].map(str) + "/" + nomes['mun_path'].map(str) + "/historico"
```

```python
html_pages = []

for idx, path in enumerate(web_paths):
    time.sleep(2)
    page = requests.get(path)
    t1 = 60
    while page.status_code == 429:
        print("Erro 249, esperando " + t1 + " segundos...")
        time.sleep(t1)
        t1 = 2 * t1
        page = requests.get(path)
    print(idx+1, " done!")
    html_pages.append(page)
```

```python
html_text = []

for idx, page in enumerate(html_pages):
    text = html_pages[idx].text
    html_text.append(text)
```

```python
import csv

with open("../data/hist_ibge_html.csv", 'wb') as myfile:
    wr = csv.writer(myfile, quoting=csv.QUOTE_ALL)
    wr.writerow(html_text)
```

```python
historicos = []

for idx, page in enumerate(html_text):
    text = BeautifulSoup(html_text[idx], 'html.parser').find("div", class_="hist__texto").get_text()
    historicos.append(text)
```
